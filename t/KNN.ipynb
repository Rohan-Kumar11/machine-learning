```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df['Species'].value_counts()
```

```python
y= df['Species']
x= df.drop('Species',axis=1)
```

```python
x.shape
```

```python
x.head()
```

```python
y.shape
```

```python
from sklearn.model_selection  import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing  import StandardScaler
sc = StandardScaler()
x_train_transformed = pd.DataFrame(sc.fit_transform(x_train),columns=x_train.columns,index=x_train.index)
```

```python
x_train_transformed.shape
```

```python
x_train_transformed.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors= 5, p=2 , metric='minkowski')
knn.fit(x_train_transformed,y_train)
```

```python
knn.classes_
```

```python
knn.effective_metric_
```

```python
knn.n_features_in_
```

```python
x_test_transformed = pd.DataFrame(sc.transform(x_test),columns=x_test.columns,index=x_test.index)
```

```python
x_test_transformed.shape
```

```python
y_test_pred = knn.predict(x_test_transformed)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test,y_test_pred)
```
